```python
lines = open('dataset/korean_dialog/hangul.txt', 'r', encoding='utf-8').readlines()
```

```python
lines
```

```python
lines = [line.split('\t')[2][:-1] for line in lines]
dialogs = [lines[i:i + 10] for i in range(0, len(lines), 10)]
```

```python
src = []
tgt = []

for dialog in dialogs:
    src += dialog[0:9]
    tgt += dialog[1:10]
```

```python
with open('dataset/korean_dialog/hangul_src.txt', 'w', encoding='utf-8') as fp:
    for s in src:
        fp.write("%s\n" % s)
        
with open('dataset/korean_dialog/hangul_tgt.txt', 'w', encoding='utf-8') as fp:
    for t in tgt:
        fp.write("%s\n" % t)
```
