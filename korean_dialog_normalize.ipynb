```python
lines = open('dataset/korean_dialog/hangul.txt', 'r', encoding='utf-8').readlines()
```

```python
lines = [line.split('\t') for line in lines]
lines
```

```python
dialogs = []
dialog = []

for i in range(len(lines)):
    if (lines[i][1] == '1' and i != 0):
        print('a')
        dialogs.append(dialog)
        dialog = []
        
    dialog.append(lines[i][2])
```

```python
dialogs
```

```python
src = []
tgt = []

for dialog in dialogs:
    src += dialog[0:9]
    tgt += dialog[1:10]
```

```python
len(tgt)
```

```python
len(src)
```

```python
with open('dataset/korean_dialog/hangul_src.txt', 'w', encoding='utf-8') as fp:
    for s in src:
        fp.write("%s\n" % s)
        
with open('dataset/korean_dialog/hangul_tgt.txt', 'w', encoding='utf-8') as fp:
    for t in tgt:
        fp.write("%s\n" % t)
```
